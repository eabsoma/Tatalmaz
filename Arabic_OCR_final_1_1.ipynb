```python
from google.colab.patches import cv2_imshow
from ArabicOcr import arabicocr
import IPython.display as ipd
import numpy
from gtts import gTTS
```

```python
pdf="/content/ميم.jpeg"
import sys, fitz
doc = fitz.open(pdf)  # open document
for page in doc:  # iterate through the pages
    pix = page.get_pixmap()  # render page to an image
    pix.save("page-%i.png" % page.number)  # store image as a PNG

```

```python
image_path='/content/ميم.jpeg'
out_image='out.jpg'
results= arabicocr.arabic_ocr(image_path,out_image)
print(results)

words=[]
for i in range(len(results)):
		word=results[i][1]
		words.append(word)
with open ('file.txt','w',encoding='utf-8')as myfile:
		myfile.write(str(words))
import cv2
img = cv2.imread('out.jpg', cv2.IMREAD_UNCHANGED)
cv2_imshow(img)
cv2.waitKey(0)
```

```python
print(words)
s = ' '.join(words)
print(s)
```

```python
pip install ar-corrector
```

```python
from ar_corrector.corrector import Corrector
corr = Corrector()

print(corr.contextual_correct(s))
```

```python
import mishkal.tashkeel
vocalizer = mishkal.tashkeel.TashkeelClass()
vocalizer.tashkeel(corr.contextual_correct(s))

```

```python
print(vocalizer.tashkeel(corr.contextual_correct(s)))
```

```python
tts = gTTS(vocalizer.tashkeel(corr.contextual_correct(s)), lang='ar')
tts.save('test.mp3')
audio_path="test.mp3"

ipd.Audio(audio_path, autoplay=True)
```
